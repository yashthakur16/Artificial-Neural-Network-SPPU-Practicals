```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# Generate input values
x_values = np.linspace(-5, 5, 100)
print(len(x_values))
# Calculate activation values for each function
sigmoid_values = sigmoid(x_values)
tanh_values = tanh(x_values)
relu_values = relu(x_values)
leaky_relu_values = leaky_relu(x_values)

# Plot the activation functions
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(x_values, sigmoid_values, label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x_values, tanh_values, label='Tanh')
plt.title('Tanh Activation Function')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x_values, relu_values, label='ReLU')
plt.title('ReLU Activation Function')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x_values, leaky_relu_values, label='Leaky ReLU')
plt.title('Leaky ReLU Activation Function')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
x_softmax = np.array([2.0, 1.0, 0.1])
plt.bar(range(len(x_softmax)), softmax(x_softmax))
plt.title('Softmax')
```
